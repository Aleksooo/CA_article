```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

# Generate some example images (replace with your actual images)
# images = [np.random.rand(100, 100) for _ in range(6)]
names = ['droplet', 'doughnut', 'worm', 'roll', 'perforation', 'layer']
images = [matplotlib.image.imread(f'/Users/alexey/Blender/img/{name}_article.png') for name in names]

# Create a figure with a 3x2 grid layout
fig, axs = plt.subplots(2, 3, figsize=(16, 8), dpi=150)

# Iterate through each subplot and plot the respective image
for i, ax in enumerate(axs.flat):
    ax.imshow(images[i], cmap='gray')  # Adjust the cmap and other parameters as needed
    ax.axis('off')  # Turn off axis labels and ticks

    # Add label 'a', 'b', 'c', 'd', 'e', 'f' in the top right corner
    ax.text(1.05, 0.2, chr(97 + i), transform=ax.transAxes,
            fontsize=28, color='black', ha='center', va='center')

# Adjust layout to prevent overlap of labels
fig.tight_layout()

# Show the plot
plt.show()
```
